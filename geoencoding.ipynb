```python
import pandas as pd
import geocoder
import os
```

```python
assets = pd.read_csv('Data/ativos.csv', sep=';')
assets.head()
```

```python
funds = pd.read_csv('Data/fundos.csv', sep=';')
funds.head()
```

```python
bing_key = os.getenv('MAPGEN_BING_KEY')
```

```python
assets['full_address'] = assets['endereco'] + ', ' + assets['cidade'] + ', ' + assets['estado']
assets.head()
```

```python
lat_long = assets['full_address'].apply(lambda address: geocoder.bing(address, key = bing_key))
lat_long
```

```python
assets['latitude'] = lat_long.apply(lambda gpos: gpos.lat)
assets['longitude'] = lat_long.apply(lambda gpos: gpos.lng)
```

```python
# Number of decimals to round to is 4
assets.round(4).to_csv('Data/assets_py.csv', index = False)
```
